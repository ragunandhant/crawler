```python
from core.crawl import get_source
from core.html2tree import HTMLParser
```

```python
url = "https://teamhub.com/blog/visualizing-project-progress-a-comprehensive-guide/"

html_code = get_source(url)
```

```python
parser = HTMLParser(html_content=html_code)
```

```python
result = parser.parse(output_file="sample.json")
```
